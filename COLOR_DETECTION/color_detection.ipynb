# Read the image

```python
img = cv2.imread("/content/drive/MyDrive/content/color_detection.jpg", 1)
cv2_imshow(img)
```

# Convert image from BGR to HSV color-space

```python
final_img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2_imshow(final_img)
```

# Create mask for color-detection

```python
lower=np.array([32,100,100])
upper=np.array([42,255,255])
area_of_interest=cv2.inRange(final_img,lower,upper)
cv2_imshow(area_of_interest)
```

# Apply mask and display result

```python
result=cv2.bitwise_and(img,img,mask=area_of_interest)
cv2_imshow(result)
```
